````python
#1) Seu código aqui

def consulta_dados(uf, tipo_interessado):
   
    resultado = df[(df['uf'] == uf) & (df['tipo_interessado'] == tipo_interessado)]
    
    if resultado.empty:
        print(f"Nenhum dado encontrado para UF {uf} e tipo de interessado {tipo_interessado}")
        return None
    
    return resultado


uf_consulta = input(' Digite a UF que quer consultar usando 2 letras maiúsculas: ')  
tipo_interessado_consulta = input(' Digite o tipo de interessado (Estado ou Município): ')  
resultado_df = consulta_dados(uf_consulta, tipo_interessado_consulta)

if resultado_df is not None:
    resultado_df = pd.DataFrame(resultado_df)
resultado_df    
````
