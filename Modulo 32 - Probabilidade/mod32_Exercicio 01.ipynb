# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
media = 15
desvpad = 0.1

# Peso a ser considerado em kg
peso = 15.2

#Utilizando a distribuio cumulativa (cdf):

probabilidade = 1 - stats.norm.cdf(peso, loc=media, scale=desvpad)

print(f'A probabilidade do pacote conter peso superior a 15.2kg  : {probabilidade.round(5)*100}%')
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Percentil de ordem 95%
percentil_95 = stats.norm.ppf(0.95, loc=media, scale=desvpad)

print(f'Percentil de 95%: {percentil_95.round(4)}kg')
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui

#Utilizando a distribuio cumulativa (cdf):

# Probabilidades para a varivel Normal Padro (Y1)
Y1_probs = stats.norm.cdf(Y)

# Probabilidades para a varivel t-student com 5 graus de liberdade (Y2)
Y2_probs = stats.t.cdf(Y, df=5)
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
plt.figure(figsize=(10, 6))
plt.plot(Y, Y1_probs, label='Normal Padro')
plt.plot(Y, Y2_probs, label='t-student (df=5)')
plt.xlabel('Valor')
plt.ylabel('Probabilidade')
plt.title('Probabilidade da Distribuio ser Inferior ao Valor:')
plt.legend()
plt.grid(True)
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu cdigo comea aqui

#Utilizando o ppf, que  o inverso da cdf:

# Percentis para a varivel aleatria Y1 (Normal Padro)
Y1_percentiles = stats.norm.ppf(Y)

# Percentis para a varivel aleatria Y2 (t-student com 5 graus de liberdade)
Y2_percentiles = stats.t.ppf(Y, df=5)
```

6. Faa um grfico com os dados do item anterior.

```python
plt.figure(figsize=(10, 6))
plt.plot(Y, Y1_percentiles, label='Normal Padro')
plt.plot(Y, Y2_percentiles, label='t-student (df=5)')
plt.xlabel('Percentil')
plt.ylabel('Valor')
plt.title('Percentis das Distribuies')
plt.legend()
plt.grid(True)
plt.show()
```
