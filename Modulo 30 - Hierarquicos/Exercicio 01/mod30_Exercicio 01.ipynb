# Mtodos hierrquicos de agrupamento - Tarefa

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

import scipy.cluster.hierarchy as shc
```

## 1) Repetindo o que fizemos em aula

Primeiramente vamos repetir o que fizemos em aula com a base de pinguins.

1. Selecione as variveis quantitativas da base de pinguins
2. Elimine valores faltantes
3. Padronize as variveis para mdia zero e desvio padro 1
4. Faa um agrupamento hierrquico dividindo essa base em 3 grupos
5. Faa o Dendrograma

```python
peng = sns.load_dataset('penguins')
peng.index.name='id'
peng_num = peng.select_dtypes(include='number').dropna()
```

```python
padronizador = StandardScaler()
peng_pad = padronizador.fit_transform(peng_num)
```

```python
clus = AgglomerativeClustering(linkage="complete", 
                                 distance_threshold = None, 
                                 n_clusters=3)

labels = clus.fit_predict(peng_pad)
```

```python
plt.figure(figsize=(15, 15))

plt.title("Penguin Dendograms")

dend = shc.dendrogram(shc.linkage(peng_pad, method='complete'))
```

## 3) Alterando o agrupamento

O Dendrogarma sugere que podemem haver grupos interessantes se solicitarmos mais que 3 grupos. Se quisermos testar o agrupamento com mais de 3 grupos, quantos grupos o Dendrograma te sugere? (Dica: no faa mais que 6 grupos).

- Analise esses agrupamentos

    - Quantos pinguins de cada espcie machos e fmeas h em cada grupo?
- Na sua opinio, o agrupamento capturou algum padro natural dos dados? Qual?

```python
from scipy.cluster.hierarchy import fcluster
```

```python
#Fazendo um dendrograma com 4 grupos:

linkage_matrix = shc.linkage(peng_pad, method='complete')

num_clusters = 4
clusters = fcluster(linkage_matrix, num_clusters, criterion='maxclust')

peng['cluster'] = pd.Series(clusters, index=peng_num.index)

# Visualizar o dendrograma 
plt.figure(figsize=(15, 15))
plt.title("Penguin Dendrogram with 4 Clusters")
dend = shc.dendrogram(linkage_matrix, color_threshold=linkage_matrix[-num_clusters + 1, 2], p=num_clusters)
plt.axhline(y=linkage_matrix[-num_clusters + 1, 2], color='r', linestyle='--')
plt.show()

# Quantos pinguins em cada grupo
cluster_counts = peng['cluster'].value_counts()
print("Nmero de pinguins em cada grupo:")
print(cluster_counts)

# Quantidade de machos e fmeas em cada grupo
gender_counts = peng.groupby(['cluster', 'sex']).size().unstack(fill_value=0)
print("Quantidade de machos e fmeas em cada grupo:")
print(gender_counts)

# Grficos de boxplot para cada varivel quantitativa
sns.set(style="whitegrid")
plt.figure(figsize=(14, 10))

for i, col in enumerate(peng_num.columns):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(x='cluster', y=col, data=peng)
    plt.title(col)

plt.tight_layout()
plt.show()
```
