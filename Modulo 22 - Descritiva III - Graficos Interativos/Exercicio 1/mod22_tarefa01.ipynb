#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
# Importando as libs:

import yfinance as yf

import plotly.express as px
from plotly import graph_objects

import pandas as pd
import numpy as npb

```

```python
ativos = ['AMZN', 'GOOG', 'MSFT']
dt_ini = '2018-01-01'  # Formato YYYY-MM-DD

yf_data = yf.download(ativos, start=dt_ini)
yf_data.head()
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
vol = yf_data['High']-yf_data['Low']
vol.head()
```

```python
vol_vert = vol.stack().reset_index()
vol_vert.columns = ['data', 'ativo', 'variacao']

vol_vert.head(100)
```

```python
fig = px.line(vol_vert, x="data", y="variacao", color='ativo', template='plotly_white')
fig.show()
```

## Com base no gráfico acima, a ação que apresentou a maior variação foi MSFT (Microsoft). Verificando a amplitude vertical mais pronunciada para a linha verde (representando a Microsoft) ao longo do tempo.


### A linha verde (MSFT) possui picos muito mais altos em relação às outras ações (AMZN e GOOG).
### Isso indica que o preço da Microsoft teve flutuações percentuais mais significativas durante o período analisado.
