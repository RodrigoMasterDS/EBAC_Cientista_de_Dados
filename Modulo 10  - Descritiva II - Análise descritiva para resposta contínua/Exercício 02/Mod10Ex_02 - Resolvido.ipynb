1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('/users/rodri/previsao_de_renda.csv')

df = df.drop(columns='Unnamed: 0').set_index(keys='index', drop=True)
df.drop_duplicates(inplace=True)
df
```

```python
plt.rc('figure', figsize=(12,12))
fig, axes = plt.subplots(2,2)

sns.pointplot(x='posse_de_imovel', 
              y='renda',  
              data=df, 
              dodge=True, 
              ax=axes[0,0])

sns.pointplot(x='posse_de_veiculo', 
              y='renda', 
              data=df, 
              dodge=True, 
              ax=axes[0,1])

sns.pointplot(x='posse_de_imovel', 
              y='renda',  
              hue='posse_de_veiculo', 
              data=df, 
              dodge=True, 
              ax=axes[1,0])

sns.pointplot(x='posse_de_veiculo', 
              y='renda', 
              hue='posse_de_imovel', 
              data=df, 
              dodge=True, 
              ax=axes[1,1])

plt.show()
```

Ao observar os gráficos acima, nota-se, na primeira linha, que ambas as variáveis qualitativas discriminam a renda de forma estatisticamente significante. No entanto, a variável posse_de_veículo apresenta maior relevância para a predição de renda, devido ao distanciamento do intervalo de confiança entre aqueles que possuem e aqueles que não possuem veículo.

Para um maior detalhamento, percebe-se, nos gráficos da segunda linha, que a variável posse_de_imóvel tem menor significância devido ao fato de que aqueles que possuem veículo têm uma renda semelhante, independentemente da posse de imóvel.
