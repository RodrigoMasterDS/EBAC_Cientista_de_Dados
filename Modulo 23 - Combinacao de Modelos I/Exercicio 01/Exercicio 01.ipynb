#### 1) Monte um passo a passo para o Bagging:

1- A partir de um dataset, realizar uma amostragem com reposição (BootStrap), de forma a montar um número desejado de conjuntos reamostrados.
2- Aplicar um modelo em cada conjunto diferente criado (os modelos podem ser iguais ou diferentes).
3- Com o resultado de cada modelo, determinar o resultado final.
4- Se o resultado for binário, fazer uma contagem de qual mais apareceu, caso seja numérico, faz-se uma média com a soma dos resultados.

#### 2) Explique com suas palavras o Bagging

Ao criar diversos conjuntos de dados reamostrados e aplicar o modelo para cada um, aproxima-se o resultado de um valor "ideal". Os conjuntos de dados são criados pela amostragem com reposição (também chamada de Bootstrap), onde se "sorteia" cada linha do novo conjunto a partir dos dados originais, onde cada linha tem todas as possibilidades independente das outras linhas. É como se os dados estivessem dentro de uma sacola, e cada um é sorteado por vez, sendo colocado de volta dentro da mesma sacola e realizando a escolha novamente.

#### 3) Implementar em python o código do Bagging

```python
import numpy as np
import random
```

```python
#Criando um conjunto de dados com valor em torno de 200 por 300x
x = np.random.normal(loc= 200.0, size=300)
x
```

```python
#Crio 4 listas de listas, através do resampling dos dados originais. Cada uma contem 300 linhas também.
list_of_lists = []
for i in range(4):
    y = random.choices(x, k=300)
    list_of_lists.append(y)
```

```python
list_of_lists
```

```python
#Calculo a média de cada conjunto (no caso seria o modelo)
medias = []
for list in list_of_lists:
    media = np.mean(list)
    medias.append(media)

medias
```

```python
#Calculo a media das médias para obter o resultado final

resultado = np.mean(medias)
resultado
```
